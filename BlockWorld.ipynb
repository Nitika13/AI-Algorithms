```python

initial=[['c','a','b'],[],[]]
goal=['a','b','c']
#current=[['c','a','b'],[],[]]
#visited=[[['c','a','b'],[],[]]]
#stack=[]

class blockworld:
  def __init__(self,initial,goal):
    self.initial=initial
    self.goal=goal
    self.visited=[]
    self.stack=None
  def states(self,state):
    for i in range(3):
      if len(state[i])>=1:
        a=state[i].pop()
        if len(state[(i+1)%3])<3:
          state[(i+1)%3].append(a)
        elif len(state[(i+2)%3])<3:
          state[(i+2)%3].append(a)
      if(state not in self.stack or state not in self.visited):
        self.stack.append(state)

  def dfs(self):
    cnt=0
    if self.stack is None:
      self.stack=list()
    self.stack.append(self.initial)
    state=self.stack.pop()
    while(True):
      self.visited.append(state)
      print(state)
      self.states(state)
      state=self.stack.pop()
      cnt+=1
      if(cnt==5):
        break
      if goal in state:
        break
    return


blockWorld = blockworld(initial,goal)
blockWorld.dfs()



```
